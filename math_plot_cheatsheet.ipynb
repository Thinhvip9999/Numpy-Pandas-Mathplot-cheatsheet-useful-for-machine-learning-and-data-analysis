# MathPlot CheatSheet

```python
# Import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Tạo dữ liệu mẫu
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

---

```python
# Câu lệnh cơ bản để vẽ đồ thị
plt.plot(x, y)  # Vẽ đồ thị x vs y
plt.title('Đồ thị Sine')  # Thêm tiêu đề
plt.xlabel('X')  # Thêm nhãn trục X
plt.ylabel('Y')  # Thêm nhãn trục Y
plt.show()  # Hiển thị đồ thị
```

---

```python
# Vẽ đồ thị với nhiều đồ thị trên cùng một biểu đồ
y2 = np.cos(x)
plt.plot(x, y, label='Sine')
plt.plot(x, y2, label='Cosine')
plt.title('Sine và Cosine')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()  # Thêm chú thích
plt.show()
```

---

```python
# Vẽ đồ thị với các kiểu đường khác nhau
plt.plot(x, y, label='Sine', color='r', linestyle='-', linewidth=2)  # Đường đỏ, đường liền
plt.plot(x, y2, label='Cosine', color='g', linestyle='--', linewidth=2)  # Đường xanh, đường đứt đoạn
plt.title('Sine và Cosine với kiểu đường khác nhau')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

---

```python
# Vẽ đồ thị với các điểm dữ liệu
plt.plot(x, y, 'bo')  # Dùng 'bo' để vẽ các điểm màu xanh
plt.title('Sine với các điểm dữ liệu')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

---

```python
# Vẽ biểu đồ cột (Bar chart)
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 8]
plt.bar(categories, values, color='orange')
plt.title('Biểu đồ cột')
plt.xlabel('Danh mục')
plt.ylabel('Giá trị')
plt.show()
```

---

```python
# Vẽ biểu đồ tròn (Pie chart)
labels = ['A', 'B', 'C', 'D']
sizes = [15, 30, 45, 10]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Biểu đồ tròn')
plt.show()
```

---

```python
# Vẽ đồ thị phân tán (Scatter plot)
x2 = np.random.rand(100)
y2 = np.random.rand(100)
plt.scatter(x2, y2, color='purple', alpha=0.6)  # alpha để điều chỉnh độ trong suốt
plt.title('Biểu đồ phân tán')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

---

```python
# Vẽ histogram (Biểu đồ tần suất)
data = np.random.randn(1000)  # Dữ liệu ngẫu nhiên phân phối chuẩn
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram')
plt.xlabel('Giá trị')
plt.ylabel('Tần suất')
plt.show()
```

---

```python
# Vẽ đồ thị với subplots (Nhiều đồ thị trên cùng một cửa sổ)
fig, axs = plt.subplots(2, 2)  # 2 hàng, 2 cột
axs[0, 0].plot(x, y)
axs[0, 0].set_title('Sine')
axs[0, 1].plot(x, y2)
axs[0, 1].set_title('Cosine')
axs[1, 0].bar(categories, values, color='green')
axs[1, 0].set_title('Bar Chart')
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
axs[1, 1].set_title('Pie Chart')
plt.tight_layout()  # Đảm bảo các đồ thị không bị chồng lên nhau
plt.show()
```

---

```python
# Lưu đồ thị vào file
plt.plot(x, y)
plt.title('Sine')
plt.xlabel('X')
plt.ylabel('Y')
plt.savefig('sine_plot.png')  # Lưu đồ thị dưới dạng file PNG
plt.close()  # Đóng đồ thị sau khi lưu
```

```python
# Vẽ đồ thị với các mảng thời gian
import matplotlib.dates as mdates
import pandas as pd

dates = pd.date_range('2022-01-01', periods=10)
values = np.random.randn(10)
plt.plot(dates, values)
plt.title('Đồ thị với thời gian')
plt.xlabel('Ngày')
plt.ylabel('Giá trị')
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.xticks(rotation=45)
plt.show()
```
